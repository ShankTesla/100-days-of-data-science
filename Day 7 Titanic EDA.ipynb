```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
titanic = pd.read_csv("train (1).csv")

# Explore the data
titanic.head()
```

```python

# Check for missing values
print(titanic.isnull().sum())

```

```python
# Visualize the distribution of passenger ages
sns.set_style("darkgrid")
plt.hist(titanic["Age"].dropna(), bins=20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Passenger Ages")
plt.show()
```

```python

# Visualize the survival rates based on gender
sns.set_style("darkgrid")
sns.countplot(x="Sex", hue="Survived", data=titanic)
plt.title("Survival Rates Based on Gender")
plt.show()
```

```python


# Visualize the survival rates based on passenger class
sns.set_style("darkgrid")
sns.countplot(x="Pclass", hue="Survived", data=titanic)
plt.title("Survival Rates Based on Passenger Class")
plt.show()
```

```python


# Visualize the survival rates based on embarkment port
sns.set_style("darkgrid")
sns.countplot(x="Embarked", hue="Survived", data=titanic)
plt.title("Survival Rates Based on Embarkment Port")
plt.show()
```

```python

# Visualize the relationship between age and fare
sns.set_style("darkgrid")
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=titanic)
plt.title("Relationship Between Age and Fare")
plt.show()
```

```python


# Explore the relationship between age and passenger class
sns.set_style("darkgrid")
sns.boxplot(x="Pclass", y="Age", data=titanic)
plt.title("Relationship Between Age and Passenger Class")
plt.show()

# Explore the relationship between fare and passenger class
sns.set_style("darkgrid")
sns.boxplot(x="Pclass", y="Fare", data=titanic)
plt.title("Relationship Between Fare and Passenger Class")
plt.show()
```

# Summary of Key Findings

## Gender and Survival

- There is a significant difference in survival rates between males and females, with females having a higher survival rate.

## Passenger Class and Survival

- There is a significant difference in survival rates between passenger classes, with first-class passengers having a higher survival rate.

## Age Distribution

- The distribution of passenger ages is slightly right-skewed, with most passengers being between 20 and 40 years old.

## Fare and Survival

- There is a positive relationship between fare and survival, with passengers who paid higher fares being more likely to survive.

## Age and Passenger Class

- There is a negative relationship between age and passenger class, with younger passengers being more likely to be in higher classes.
